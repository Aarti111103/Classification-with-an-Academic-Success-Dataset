# 5.  Model Training

Logistic Regression Model For Multiclass Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Separate features and target
X = df_train.drop('Target', axis=1)
y = df_train['Target']

# Encode the target variable if it's categorical
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize and fit the scaler on the training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train a logistic regression model for multiclass classification
lr = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model = lr.fit(X_train_scaled, y_train)

# Make predictions on the test set
predictions = model.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
```

```python
# Print classification report for detailed evaluation
print(classification_report(y_test, predictions, target_names=label_encoder.classes_))
```

```python
# Use the model for new predictions
# df_test is testing dataset

# First column in Train dataset is the student ID
student_ids = df_test.iloc[:, 0]

# Drop the student ID column for scaling and prediction
new_test_data_features = df_test.drop(df_test.columns[0], axis=1)

# Ensure that new_test_data_features has the same feature columns as X_train
missing_cols = set(X_train.columns) - set(new_test_data_features.columns)
for col in missing_cols:
    new_test_data_features[col] = 0

new_test_data_features = new_test_data_features[X_train.columns]

# Scale the new test data using the same scaler
new_test_data_scaled = scaler.transform(new_test_data_features)

# Make predictions on the new test data
new_predictions = model.predict(new_test_data_scaled)

# Convert numerical predictions back to original labels
new_predictions_labels = label_encoder.inverse_transform(new_predictions)

# Create a DataFrame with student IDs and their corresponding predictions
predictions_df = pd.DataFrame({'Student ID': student_ids, 'Prediction': new_predictions_labels})

# Display the predictions
print(predictions_df)
```
